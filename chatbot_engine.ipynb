## 📝 **Executive Summary**
This notebook explores the development of a retrieval-based chatbot using Natural Language Processing (NLP). It documents the transition from raw text data to a searchable corpus using mathematical similarity.


## 🎯 **Project Overview & Objective**
The goal of this project is to build a reliable system that finds the best answer from a specific text file. This chatbot works by looking for exact word matches between a user's question and the information it has been given.

**How it works:**
1.  **Preparation:** The system reads a text file and breaks it down into individual sentences.

2.  **Cleaning:** It removes common "filler" words (like 'the' or 'is') to focus only on the keywords that carry the most meaning.

3.  **Matching:** When a question is asked, the bot compares the keywords in that question against every sentence in its memory.

4.  **Scoring & Retrieval:** The bot calculates a confidence score based on word overlap (Jaccard Similarity). It only selects and displays the sentence with the highest score if that score is above zero, ensuring the answer is relevant.

By following this logic, the chatbot avoids "hallucinating" or making up incorrect facts, ensuring that every response is grounded in the provided data.

## **1. Environment Setup & Libraries**

This section imports Python's built-in tools for handling text and downloads specific "dictionaries" (NLTK data) that help the computer understand the structure of the English language.

```python
# Imports the necessary libraries 
import nltk
import string
import warnings
import pickle
import os
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Suppressing warnings for a cleaner output
warnings.filterwarnings("ignore")

# NLTK data downloads
nltk.download("punkt", quiet=True)
nltk.download("punkt_tab", quiet=True) 
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)

print("Workspace is ready and all tools are loaded.")
```

## **2. Data Acquisition & Initial Segmentation**

This section opens a text file (chatbot.txt), reads the information inside, and breaks it down into individual sentences. This allows the bot to look through the file sentence-by-sentence to find an answer.

```python
# Create the full path to the chatbot.txt file
file_path = os.path.join(r"C:\Users\pc\Desktop\GoMyCode\chatbot_engine", "chatbot.txt")

# Open the file and read the text
try:
    with open(file_path, "r", encoding="latin-1") as file:
        raw_content = file.read().replace("\n", " ")
except FileNotFoundError:
    #  Fallback text used if the file is missing from the directory
    raw_content = "File not found. Please ensure 'chatbot.txt' is in the folder."

# Split the big block of text into a list of single sentences
sentences = sent_tokenize(raw_content)

print(f"Successfully read the file. The bot has {len(sentences)} sentences to learn from.")
```

## **3. Cleaning the Text**

To make the text searchable, it must be standardised. This function converts text to lowercase, removes punctuation and common stopwords (like 'the' or 'is'), and applies lemmatisation to reduce words to their dictionary root. This ensures that words such as "chatting" and "chat" are treated as the same term.

```python
# Tools to simplify words (e.g. changing 'running' to 'run')
lemmatizer = WordNetLemmatizer()
common_filler_words = set(stopwords.words("english"))

def clean_sentence(text):
    # Break the sentence into words
    words = word_tokenize(text)
    
    # Keep only the important words and simplify them
    important_words = [
        lemmatizer.lemmatize(w.lower()) 
        for w in words 
        if w.lower() not in common_filler_words and w not in string.punctuation
    ]
    return important_words

# Clean every sentence in the list to ensure the sentences are searchable for the bot
cleaned_corpus = [clean_sentence(s) for s in sentences]

print("Text cleaning is complete.")
```

## **4. Finding the Best Match**

When a user asks a question, this section compares the words in that question to the words in every sentence of the text file. It calculates a similarity score and picks the sentence that has the most words in common with the user's question.

```python
def find_answer(user_question):
    # Clean the user's question first
    question_words = clean_sentence(user_question)
    
    if not question_words:
        # Returning two values - response and score
        return "Please enter a valid question.", 0.0
        
    best_score = -1
    best_index = 0
    
    for i, sentence_words in enumerate(cleaned_corpus):
        # Find which words the question and the sentence share (intersection)
        shared_words = set(question_words).intersection(sentence_words)
        
        # Find all unique words found in the question and the sentence combined (union)
        all_words = set(question_words).union(sentence_words)
                
        # Calculate Jaccard Similarity score (0.0 to 1.0)
        score = len(shared_words) / float(len(all_words)) if all_words else 0

        if score > best_score:
            best_score = score
            best_index = i

    # Return the answer and the score (rounded)
    if best_score > 0:
        return sentences[best_index], round(best_score, 2)
    else:
        return "I'm sorry, I couldn't find a relevant answer.", 0.0

print("The matching logic is ready. The bot can now compare questions to answers.")
```

## **5. Testing the Chatbot**

In this final step, everything is brought together. A simple function is created to talk to the bot, and a test question is run to ensure it returns a helpful response. This proves the system is working from start to finish.

```python
def start_chatting(text_input):
     # This captures two values from the matching function: answer and score
    return find_answer(text_input)

# List of questions to test the bot's accuracy
test_questions = [
    "What is a chatterbot?",
    "Tell me about messenger.",
    "How does a chatbot help customers?",
    "Can a chatbot simulate human conversation?",
    "Tell me about chatterbots",
    "How is a chatterbot used for customer service?",
    "Can you explain the meaning of an automated conversation?",
    ]

# Validation test
print("--- Start Validation Tests ---")

for test_input in test_questions:
    # Unpacking the two returned values: response and score
    bot_response, confidence_score = start_chatting(test_input)
    
    print(f"\nUser asked: {test_input}")
    print(f"Confidence Score: {confidence_score}")
    print(f"Bot replied: {bot_response}")
    print("-" * 50)
```

## **6. Saving the Model**

This section ensures the chatbot is fast and efficient by saving the processed knowledge base into a specialised file format known as a pickle file (.pkl). Instead of requiring the computer to re-read and clean the entire text file during every startup, the finished work is "frozen" so it can be reloaded instantly by the Streamlit app. This conserves time and computational power.

```python
# Define the name of the saved model file
project_path = r"C:\Users\pc\Desktop\GoMyCode\chatbot_engine"

# Define the full name and location for the saved model file
save_file = os.path.join(project_path, "chatbot_model.pkl")

# Save the original sentences and cleaned corpus for the App to use
try:
    with open(save_file, "wb") as f:
        pickle.dump((sentences, cleaned_corpus), f)
    print(f"Model successfully saved as 'chatbot_model.pkl' in {project_path}")
except Exception as e:
    print(f"An error occurred while saving: {e}")
```

## 📖 **Glossary**
*   **Tokenisation:** Breaking a paragraph into single sentences or words.

*   **Intersection:** In the context of this bot, it refers to the set of keywords that are found in both the user's question and a sentence from the text file.

*   **Union:** This is the total collection of unique keywords found in both the user's question and a sentence combined.

*   **Tokenisation:** Breaking a paragraph into single sentences or words.

*   **NLTK (Natural Language Toolkit):** A popular Python library used to help computers understand, analyse, and process human language.

*   **UTF-8 encoding:** The most common international standard for saving text files so that computers can read them correctly.

*   **Latin-1:** An alternative way of reading text files that is better at handling special characters (like smart quotes or dashes) found in some Windows documents.

*   **UnicodeDecodeError:** An error that happens when a computer tries to read a text file using the wrong "language" or encoding rules.

*   **Jaccard Similarity:** A mathematical way to measure how similar two sentences are by looking at how many words they have in common.

*   **Model:** A set of instructions or a saved file that a computer uses to perform a specific task, like answering questions.

*   **Corpus:** A technical term for a large collection of text data used to inform an AI.

*   **Lemmatisation:** Changing a word back to its simplest form (e.g., "talking" becomes "talk").

*   **Stopwords:** Common words (like "and", "the", "a") that the bot ignores to save time.

*   **Data Persistence:** The process of saving work to a file so that it is not lost when the program is closed.

*   **Overlap:** In the context of a chatbot, this refers to the specific words that the user's question and a sentence in the text file have in common. The more words that match or "overlap," the higher the confidence score will be.
